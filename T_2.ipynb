**For Printing And Making File**

```python
from PyPDF2 import PdfReader

def extract_text_from_pdf(pdf_path, output_text_file):
    reader = PdfReader(pdf_path)
    text = ""

    # Concatenate text from all pages
    for page_num in range(len(reader.pages)):
        page = reader.pages[page_num]
        text += page.extract_text()

    # Write the extracted text to a text file
    with open(output_text_file, "w", encoding="utf-8") as text_file:
        text_file.write(text)

# Specify the paths
pdf_path = "/content/PDF/&supremecourt&2011&18235&18235_2011_5_1501_36148_Judgement_14-Jun-2022.pdf"
output_text_file = "/content/TXT/&supremecourt&2011&18235&18235_2011_5_1501_36148_Judgement_14-Jun-2022_extracted.txt"

# Extract text and write to a text file
extract_text_from_pdf(pdf_path, output_text_file)
```

# Processing TEXT File

```python
from opennyai import Pipeline
from opennyai.utils import Data

# Specify the local file paths
file_path1 = '/content/TXT/&supremecourt&2011&18235&18235_2011_5_1501_36148_Judgement_14-Jun-2022_extracted.txt'
file_path2 = '/content/TXT/&supremecourt&2011&18235&18235_2011_5_1501_36148_Judgement_14-Jun-2022_extracted.txt'

# Read the content of the local text files
with open(file_path1, 'r', encoding='utf-8') as file1:
    text1 = file1.read()

with open(file_path2, 'r', encoding='utf-8') as file2:
    text2 = file2.read()

# Create a Data object for data preprocessing before running ML models
texts_to_process = [text1, text2]
data = Data(texts_to_process)

# If you have access to GPU, set this to True; otherwise, set it to False
use_gpu = True

# Choose which of the AI models you want to run from the 3 models 'NER', 'Rhetorical_Role','Summarizer'.
# For example, if just Named Entity is of interest, then just select 'NER'
pipeline = Pipeline(components=['NER', 'Rhetorical_Role', 'Summarizer'], use_gpu=use_gpu, verbose=True)

results = pipeline(data)
```

Removes Page Number **BUT ALL NUMBER GOT DELETED**

```python
from PyPDF2 import PdfReader

def extract_text_from_pdf(pdf_path, output_text_file):
    reader = PdfReader(pdf_path)
    text = ""

    for page_num in range(len(reader.pages)):
        page = reader.pages[page_num]
        page_text = page.extract_text()

        # Remove lines that are likely to contain page numbers
        lines = page_text.split('\n')
        cleaned_lines = [line for line in lines if not any(c.isdigit() for c in line.strip())]

        # Concatenate cleaned lines
        cleaned_page_text = '\n'.join(cleaned_lines)
        text += cleaned_page_text

    # Write the extracted text to a text file
    with open(output_text_file, "w", encoding="utf-8") as text_file:
        text_file.write(text)

# Specify the paths
pdf_path = "/content/&judgment&judis&43486.pdf"
output_text_file = "/content/&judgment&judis&43486_extracted.txt"

# Extract text and write to a text file
extract_text_from_pdf(pdf_path, output_text_file)
```
